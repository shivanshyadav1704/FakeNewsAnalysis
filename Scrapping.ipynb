```python
import pandas as pd
from ntscraper import Nitter
```

```python
scraper=Nitter(0)
```

```python
tweets = scraper.get_tweets('8bit__goldy', mode = 'user', number=10)
```

```python
tweets
```

```python
final_tweets = []
for x in tweets['tweets']:
    data = [x['link'], x['text'],x['date'],x['stats']['likes'],x['stats']['comments']]
    final_tweets.append(data)
```

```python
dat = pd.DataFrame(final_tweets, columns =['twitter_link','text','date','likes','comments'])
```

```python
from confluent_kafka import Producer
```

```python
# Kafka bootstrap servers
bootstrap_servers = 'localhost:9092'

# Create Kafka producer configuration


conf = {'bootstrap.servers': bootstrap_servers}

# Create Kafka producer instance
producer = Producer(conf)

topic = 'twitter'
```

```python
 for tweet in tweets['tweets']:
        producer.produce(topic, value=tweet['text'].encode('utf-8'))

producer.flush()
```
